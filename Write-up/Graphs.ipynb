```python
import matplotlib.pyplot as plt
from matplotlib.patches import Ellipse

# Create figure with 2 panels
fig, axes = plt.subplots(1, 2, figsize=(10, 5))
fig.suptitle("Figure 1: Schematic Representation of Cardiac Morphology in HCM and DCM",
             fontsize=12, weight='bold')

# -------- Panel A: HCM --------
ax = axes[0]
ax.set_title("Panel A: Hypertrophic Cardiomyopathy (HCM)", fontsize=10)
ax.set_xlim(0, 10)
ax.set_ylim(0, 10)
ax.axis('off')

# Thick ventricular wall (outer ellipse)
outer_hcm = Ellipse((5, 5), width=6, height=8, edgecolor='black', facecolor='#cce5ff', lw=2)
ax.add_patch(outer_hcm)

# Small LV cavity (inner ellipse)
inner_hcm = Ellipse((5, 5), width=2, height=4, edgecolor='black', facecolor='white', lw=2)
ax.add_patch(inner_hcm)

# LVOT obstruction arrow
ax.annotate("LVOT Obstruction", xy=(6.5, 6.5), xytext=(8.5, 8),
            arrowprops=dict(arrowstyle="->", lw=1.5), fontsize=8)

# -------- Panel B: DCM --------
ax = axes[1]
ax.set_title("Panel B: Dilated Cardiomyopathy (DCM)", fontsize=10)
ax.set_xlim(0, 10)
ax.set_ylim(0, 10)
ax.axis('off')

# Thinned ventricular wall (outer ellipse)
outer_dcm = Ellipse((5, 5), width=8, height=10, edgecolor='black', facecolor='#f8d7da', lw=2)
ax.add_patch(outer_dcm)

# Enlarged LV cavity (inner ellipse)
inner_dcm = Ellipse((5, 5), width=6, height=8, edgecolor='black', facecolor='white', lw=2)
ax.add_patch(inner_dcm)

# Reduced contractility annotation
ax.annotate("Reduced Contractility", xy=(7, 5), xytext=(8.5, 7),
            arrowprops=dict(arrowstyle="->", lw=1.5), fontsize=8)

plt.tight_layout()
plt.show()
```
